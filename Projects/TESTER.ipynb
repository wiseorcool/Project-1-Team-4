```python
response = requests.get(base_url).json()
print(json.dumps(response, indent=4, sort_keys=True))
```

```python
data_df = pd.DataFrame(response)
data_df.count()
```

```python
headers = ['date_rptd','date_occ','area_name','crm_cd', 'vict_age', 'vict_sex','vict_descent', 'lat','lon','weapons_used_cd']
data_clean = data_df.reindex(columns=headers)
data_clean.head()
```

```python
data_clean['Year'] = pd.to_datetime(data_clean['date_occ']).dt.strftime('%Y')
data_cleaned = data_clean.drop(['date_occ'], axis=1)
data_cleaned.head()
#year = year of occurrence 
```

```python
data_cleaned.to_csv('woot.csv',index=False)
```
